```python
import glob
import os
import pathlib
import zipfile
import rarfile
import time
import shutil
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
os.chdir(os.getcwd())

counter = 0
task_folder = '../Task 1/Submission/'
for folder, dirs, files in os.walk(task_folder):
    counter += 1
    for zip_file_name in files:
        if(len(files) > 0):
            zip_file_path = folder + '\\' + zip_file_name
            id_mhs = zip_file_name.find('.zip')
            id_mhs = zip_file_name[:id_mhs]
            print(id_mhs)

            root_dest = '../Task 1/Submission_src/'
            rar = False
            filename, file_extension = os.path.splitext(zip_file_path)
            if file_extension == '.zip':
                archive = zipfile.ZipFile(zip_file_path)
            else:
                print('rar at ' + zip_file_path)
                continue

            base = os.path.basename(filename)
            print('base ' + base)
            for file in archive.namelist():
                if file.endswith('.java') and not os.path.basename(file).startswith('._'):
                    fn = os.path.basename(file)
                    dest_path = root_dest + id_mhs

                    if not os.path.exists(os.path.join(os.getcwd(), dest_path)):
                        os.makedirs(os.path.join(os.getcwd(), dest_path))

                    print('processing ' + file)
                    source = archive.open(file)
                    target = open(dest_path + '/' + fn, "wb+")
                    with source, target:
                        shutil.copyfileobj(source, target)

            archive.close()
    #         time.sleep(0.1)
            print('src extracted at ' + dest_path + ' from ' + zip_file_path)
            print()
```

```python
src_root2 = '../Task 1/Submission_src/'
folders_to_check2 = os.listdir(src_root2)
folders_to_check2
```

```python
for i in range(len(folders_to_check2)):
    if i < len(folders_to_check2) - 1:
        for j in range(i + 1, len(folders_to_check2)):
            folder_1 = [src_root2 + dir_name + '/' + fn for dir_name in [folders_to_check2[i], folders_to_check2[j]] for fn in os.listdir(src_root2 + dir_name)]
            documents = [open(f).read() for f in folder_1]
            tfidf = TfidfVectorizer().fit_transform(documents) # konversi dokumen ke vektor
            cosine_sim = cosine_similarity(tfidf, tfidf)
            counter = 0
            f = [False] * len(folder_1)
            flag = [f] * len(folder_1)
            for row in cosine_sim:
                for item in row:
                    if item > 0.85 and item <= 0.9999999999999:
                        index_x, index_y = np.where(cosine_sim == item)
                        if(os.path.dirname(folder_1[index_x[0]]) != os.path.dirname(folder_1[index_y[0]]) and not flag[index_x[0]][index_y[0]]):
                            flag[index_x[0]][index_y[0]] = True
                            print('Score: %3f file1: %s file2: %s' % (item, folder_1[index_x[0]], folder_1[index_y[0]]))
                counter += 1    
```
